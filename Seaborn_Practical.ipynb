```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style('dark', {'axes.facecolor': 'black'})

# Create figure and axes
f, axes = plt.subplots(2, 2, figsize=(15, 15))

# KDE Plot [0,0] - Budget vs Audience Rating
k1 = sns.kdeplot(x=df.BudgetMillions, y=df.AudienceRating, fill=True, cmap='inferno', ax=axes[0, 0], thresh=0)
k1b = sns.kdeplot(x=df.BudgetMillions, y=df.AudienceRating, cmap='cool', ax=axes[0, 0])

# KDE Plot [0,1] - Budget vs Critic Rating
k2 = sns.kdeplot(x=df.BudgetMillions, y=df.CriticRating, fill=True, cmap='inferno', ax=axes[0, 1], thresh=0)
k2b = sns.kdeplot(x=df.BudgetMillions, y=df.CriticRating, cmap='cool', ax=axes[0, 1])

# Violin Plot [1,0] - Drama Genre: Years vs Critic Rating
z = sns.violinplot(data=df[df.Genre == 'Drama'], x='Years', y='CriticRating', ax=axes[1, 0])

# KDE Plot [1,1] - Critic Rating vs Audience Rating
k4 = sns.kdeplot(x=df.CriticRating, y=df.AudienceRating, fill=True, cmap='Blues_r', ax=axes[1, 1], thresh=0)
k4b = sns.kdeplot(x=df.CriticRating, y=df.AudienceRating, cmap='gist_gray_r', ax=axes[1, 1])

# Set x-limits
axes[0, 0].set_xlim(-20, 160)
axes[0, 1].set_xlim(-20, 160)

# Show plot
plt.show()
```
