```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load iris dataset using seaborn
iris = sns.load_dataset('iris')

# Drop the non-numeric column ("species") for correlation computation
iris_numeric = iris.drop(columns=['species'])

# Create figure
fig, ax = plt.subplots(figsize=(10, 7))

# Heatmap
sns.heatmap(
    iris_numeric.corr(),  # No need for numeric_only=True as non-numeric column is removed
    annot=True,
    cmap='cubehelix',
    linewidths=1,
    linecolor='k',
    square=True,
    vmin=-1,
    vmax=1,
    cbar=True,
    cbar_kws={"orientation": "vertical"},
    ax=ax  # Assign heatmap to the axis
)

# Show plot
plt.show()
```

```python

# Filter for Setosa species
sub = iris[iris['Species'] == 'Iris-setosa']

# Create KDE plot
plt.figure(figsize=(8, 6))
sns.kdeplot(
    x=sub['SepalLengthCm'], 
    y=sub['SepalWidthCm'], 
    cmap="plasma", 
    fill=True,  # Use fill instead of shade
    thresh=0.05  # Adjust threshold instead of shade_lowest
)

# Labels and title
plt.title('Iris-setosa')
plt.xlabel('Sepal Length (Cm)')
plt.ylabel('Sepal Width (Cm)')

# Show plot
plt.show()
```
